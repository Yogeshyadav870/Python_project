# Objective 3 : Trend of Movie Production over the years

* Count Movies per Year
* Bar Plot of Yearly Releases



```python
movies_per_year = df['Year'].value_counts().sort_index()
```

```python
sns.lineplot(x=movies_per_year.index, y=movies_per_year.values)
plt.title("Movies Released per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
```

Bar plot of yearly releases ->


* Clearly visualize high and low production years



```python
sns.barplot(x=movies_per_year.index, y=movies_per_year.values)
plt.xticks(rotation=45)
```

#### Identify Peak & Low Years ->

* Purpose: Lable years with most and fewest releases



```python
peak_year = movies_per_year.idxmax()
low_year = movies_per_year.idxmin()
print(f"Peak Year: {peak_year} with {movies_per_year.max()} movies")
print(f"Lowest Year: {low_year} with {movies_per_year.min()} movies")
```

# Objective 4: Genre Based Analysis


* Handle Multiple Genres
* Most Frequent Genres



```python
df_genre = df.copy()
df_genre['Genre'] = df_genre['Genre'].str.split(',')
df_genre = df_genre.explode('Genre')
df_genre['Genre'] = df_genre['Genre'].str.strip()
```

```python
genre_counts = df_genre['Genre'].value_counts()
genre_counts.plot(kind='bar', figsize=(12, 6), title='Top Genres by Frequency')
```

Average Rating per Genre ->


* Find out which genres are generally more liked



```python
avg_rating = df_genre.groupby('Genre')['Rating'].mean().sort_values(ascending=False)
avg_rating.plot(kind='barh', title='Average IMDb Rating by Genre')
```

#### Average Revenue per Genre->


* purpose: Identify commercially successful genres





```python
avg_revenue = df_genre.groupby('Genre')['Revenue (Millions)'].mean().sort_values(ascending=False)
avg_revenue.plot(kind='barh', title='Average Revenue by Genre')
```

# Objective 5: Director-Based Performance Analysis


* Check for Missing Values in 'Director' Column



```python
# cleaning the data first
df['Director'].isnull().sum()
df['Director'].dropna(inplace=True)
```

#### Top directors by no. of movies ->

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_directors = df['Director'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values,
            y=top_directors.index,
            palette="cividis")

plt.title('Top 10 Directors by Number of Movies', fontsize=16)
plt.xlabel('Number of Movies', fontsize=12)
plt.ylabel('Director', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt
```

```python
top_directors = df['Director'].value_counts().head(10)
top_directors.plot(kind='barh', title='Top 10 Directors by Number of Movies')
```

#### Average rating per Director ->


* purpose: Discover directors known for highly-rated movies



```python
avg_rating = df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(10)
avg_rating.plot(kind='barh', title='Top 10 Directors by Average Rating')
```

#### Top Directors with Most High-Rated Movies ->


* purpose: high-rated movies with a rating above,says 8.0



```python
high_rated = df[df['Rating'] >= 8.0]
top_high_rated_directors = high_rated['Director'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_high_rated_directors.values,
            y=top_high_rated_directors.index,
            palette="viridis")  # Use a color palette

plt.title('Directors with Most Highly Rated Movies (Rating >= 8.0)', fontsize=16)
plt.xlabel('Number of Highly Rated Movies', fontsize=12)
plt.ylabel('Director', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add a subtle grid

plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
high_rated = df[df['Rating'] >= 8.0]
top_high_rated_directors = high_rated['Director'].value_counts().head(10)
top_high_rated_directors.plot(kind='bar', title='Directors with Most Highly Rated Movies')
```

# Objective 7 :
###Identify relationships and correlations b/t num. variables like Rating, Votes, Revenue, Runtime and Metascore

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = ['Rating', 'Votes', 'Revenue (Millions)', 'Runtime (Minutes)', 'Metascore']
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

Pair plot ->

```python
sns.pairplot(df[numeric_cols])
```
